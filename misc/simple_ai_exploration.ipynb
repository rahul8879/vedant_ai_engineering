```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# lets generate retail sales data for a year
dates = pd.date_range(start='2023-01-01', end='2023-12-31', freq='D')
sales = np.random.randint(100, 1000, size=len(dates))
retail_data = pd.DataFrame({'Date': dates, 'Sales': sales})
retail_data.head()
```

```python
# plot the sales data
plt.figure(figsize=(12, 6))
plt.plot(retail_data['Date'], retail_data['Sales'], marker='o', linestyle='-')
plt.title('Daily Retail Sales in 2023')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.show()
```
