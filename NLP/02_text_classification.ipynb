```python
import pandas as pd
from sklearn.datasets import fetch_20newsgroups

# convert
#  the data into pandas dataframe
data = fetch_20newsgroups(subset='all')
df = pd.DataFrame({'text': data.data, 'target': data.target})
df.head(2)
```

```python
# we can build neural network models for text classification using this dataframe
# pytorch : 
import torch

# vectorizers and tokenizers

from transformers import AutoTokenizer

tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")
tokens = tokenizer(
    "This is a sample text for tokenization",
    padding="max_length",
    truncation=True,
    max_length=32,
    return_tensors="pt",
)


# embedding layers
from transformers import AutoModel
model = AutoModel.from_pretrained("bert-base-uncased")
embeddings = model(**tokens).last_hidden_state
embeddings.shape  # (batch_size, sequence_length, embedding_dim)
```
